```python
#Sigmoid activation function
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10,10)
y = []
for i in range (x.shape[0]):
  y.append((1/(1 + np.exp(-x[i]))))
plt.plot(x,y,c="red")
plt.title("Sigmoid")
plt.show()
```

```python
#Hyperbolic Tangent activation function
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10,10)
y = []
for i in range (x.shape[0]):
  y.append( (np.exp(x[i])-np.exp(-x[i]))/(np.exp(x[i])+np.exp(-x[i])))
plt.plot(x,y,c="red")
plt.title("Hyperbolic Tangent")
plt.show()
```

```python
#Rectified Linear Unit activation function
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10,10)
y = []
for i in range (x.shape[0]):
  y.append( max (0,x[i]))
plt.plot(x,y,c="red")
plt.title("ReLU")
plt.show()
```
