```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('spam.csv')
df
```

```python
df.groupby('Category').describe()
```

```python
# we have to convert category and message column both in numbers
df['Category_in_binary'] = df['Category'].apply(lambda x:1 if x == 'spam' else 0 )
df.head(6)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.Message,df.Category_in_binary,test_size=0.25,random_state=42)
```

```python
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)

X_train_count.toarray()[:2]
```

```python
X_test = [str(x) for x in X_test]
X_test_count = v.transform(X_test)
```

```python
model = MultinomialNB()
X_train_count.shape
```

```python
y_train.shape
```

```python
model.fit(X_train_count,y_train)
```

```python
predictions = model.predict(X_test_count)
```

```python
accuracy = accuracy_score(y_test, predictions)
accuracy
```

```python
new_emails = ["Free Viagra!", "Meeting tomorrow at 10 AM.", "Important document attached."]
new_emails_count = v.transform(new_emails)
predictions = model.predict(new_emails_count)
```

```python
for email, prediction in zip(new_emails, predictions):
    print(f"Email: '{email}' - Predicted Label: {'Spam' if prediction == 1 else 'Not Spam'}")
```
