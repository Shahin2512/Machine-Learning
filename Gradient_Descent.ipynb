```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data-gd.csv')
```

```python
x = data.iloc [:,0]
y = data.iloc [:,1]
plt.scatter(x,y)
plt.show()
```

```python
# building model
m = 0
c = 0
learning_rate = 0.0001
epochs = 1000
n = float(len(x))
for i in range (len(x)):
  prediction = m * x + c
  D_m = (-2/n) * sum (x * (y - prediction))
  D_c = (-2/n) * sum (y - prediction)
  m = m - learning_rate * D_m
  c = c - learning_rate * D_c
print(m,c)
```

```python
# prediction
prediction = m * x + c
```

```python
plt.scatter (x,y)
plt.plot ([min(x),max(x)],[min(prediction),max(prediction)], color='red')
plt.show()
```
