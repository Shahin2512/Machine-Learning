```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv', encoding='latin-1')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().any()
```

```python
sns.heatmap(df.isnull(), yticklabels=False,cbar=False , cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# Combining 2 data frames
# how = 'left' (focus on left table)
final_df=pd.merge(df,df_country, on ='Country Code' , how='left')
```

```python
final_df.head(2)
```

```python
# To check data types
final_df.dtypes
```

```python
# observing data
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
# pie chart for top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3], autopct='%1.2f%%')
```

Observation: Zomat maximum records or transaction are from India.
After that US and then United Kingdom

```python
final_df.columns
```

```python
# Zomato ratings
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings.head(5)
```

## Observation
 1. Ratings between 4.5 to 4.9 ---> Excellect
 2. Ratings between 4.0 to 3.4 ---> Very Good
 3. Ratings between 3.5 to 3.9 ---> Good
 4. Ratings between 3.0 to 3.4 ---> Average
 5. Ratings between 2.5 to 2.9 ---> Average
 6. Ratings between 2.0 to 2.4 ---> Poor

```python

matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color", data=ratings, palette=['white','red','orange','yellow','green','green'])
```
